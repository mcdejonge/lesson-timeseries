# Processing California Earthquake Data

This notebook contains the processing for turning the raw "combined_eq_california_clean.csv" into a time series listing number of hours since last earthquake, magnitude of last earthquake and magnitude of earthquake.

It is provided here not only in the interest of transparency but also because it contains some techniques you might find useful in your own work.

```python
# import libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from statsmodels.graphics import tsaplots
from statsmodels.tsa.stattools import acf
import statsmodels.tsa.api as tsa
from scipy.signal import savgol_filter, butter, filtfilt
from statsmodels.tsa.seasonal import seasonal_decompose
from sklearn.metrics import mean_absolute_error, mean_absolute_percentage_error
```

```python
# Cleanup
raw = pd.read_csv("data/combined_eq_california_clean.csv")[['type', 'time', 'mag', 'place']]

raw = raw[raw['type'] == 'earthquake'][['time', 'mag', 'place']]
raw['time'] = pd.to_datetime(raw['time'])
raw = raw[raw['time'] > '2015-01-01'].reset_index(drop=True) # Reduce the amount of data to avoid long processing times.
# plt.plot(raw['time'], raw['mag'])
# plt.show()
# California only
raw = raw[(raw['place'].str.contains('California')) | raw['place'].str.contains(', CA')]
raw = raw.sort_values(['time']).reset_index(drop=True)
# raw.set_index(['time'])

raw.resample(rule='60Min', on='time').max().dropna().reset_index()
```
